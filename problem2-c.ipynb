```
#Problem 2 - (c)
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import *
ax = 1.5;
bx = 5.0;
Zx = 0.4;
nzx = 2.7;
Xz = 1.5;
nxz = 2.7;
az = 1;

m = 300;
X = np.linspace(0.2, 3.6, m);
S = np.zeros(m); #E[0]
Z = np.zeros(m); #E[1]
for q in range(m):
    space = np.array([1,1])
    def SSequation(E):
        W = np.empty((2))
        W[0] = (ax + bx * E[0])/(1 + E[0] + np.power(E[1]/Zx, nzx)) - X[q]
        W[1] = az/(1 + np.power(X[q]/Xz, nxz)) - E[1]
        return W
    S[q],Z[q] = fsolve(SSequation, space)
    space[0] = S[q];
    space[1] = Z[q];
    
plt.plot(S, X ,'r');
plt.title('Gene expression vs. signal S');
plt.xlabel('S');
plt.ylabel('Gene expression X');
plt.show()
```
